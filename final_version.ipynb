```python
import random

def mastermind_attempt(choice):
    """In the mastermind game, the player has to guess a secret combination of color options.
    
    This function helps determine if there are some right colors and if they are correctly located in the list,
    by comparing the player attempt to the exact solution.
    
    Input: List of choice among color options
    Output: Answers to matching color or color&placement in the secret combination"""
    
    global solution
    coder_answers = []
    copy_choice = choice.copy()
    copy_solution = solution.copy()

    for i in range(len(choice)):
        if copy_choice[i] == copy_solution[i]:
            coder_answers.append('RightColor&Placement')
            copy_choice[i] = 'a'
            copy_solution[i] = 'b'
        else:
            continue
    for i in range(len(choice)):
        if copy_choice[i] in copy_solution:
            copy_solution[copy_solution.index(copy_choice[i])] = 'c'
            copy_choice[i] = 'd'
            coder_answers.append('RightColor')
        else:
            continue
            
    return coder_answers

```

```python
class PrintInColor:
    
    """This class is created to write some sentences in color or bold or underline"""
    
    RED = '\033[91m'
    GREEN = '\033[92m'
    YELLOW = '\033[33m'
    WHITE = '\33[37m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    BLACK= '\033[0m'
    END = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    INVISIBLE = '\033[08m'

    @classmethod
    def red(cls, s, **kwargs):
        print(cls.RED + cls.BOLD + s + cls.END, **kwargs)

    @classmethod
    def green(cls, s, **kwargs):
        print(cls.GREEN + cls.BOLD + s + cls.END, **kwargs)

    @classmethod
    def yellow(cls, s, **kwargs):
        print(cls.YELLOW + cls.BOLD + s + cls.END, **kwargs)

    @classmethod
    def blue(cls, s, **kwargs):
        print(cls.BLUE + cls.BOLD + s + cls.END, **kwargs)

    @classmethod
    def white(cls, s, **kwargs):
        print(cls.WHITE + cls.BOLD+ s + cls.END, **kwargs)
        
    @classmethod
    def black(cls, s, **kwargs):
        print(cls.BLACK + cls.BOLD + s + cls.END, **kwargs)

    @classmethod
    def bold(cls, s, **kwargs):
        print(cls.BOLD + s + cls.END, **kwargs)
        
    @classmethod
    def underline(cls, s, **kwargs):
        print(cls.UNDERLINE + cls.BOLD + s + cls.END, **kwargs)
```

```python
def color_in_color(colors):
    
    """This function is created in order to write a list of color in their color"""
    
    for i in range(len(colors)):
        if colors[i]=='Yellow':
            PrintInColor.yellow('Yellow', end=' ')
        if colors[i]=='Red':
            PrintInColor.red('Red', end=' ')
        if colors[i]=='Blue':
            PrintInColor.blue('Blue', end=' ')
        if colors[i]=='Black':
            PrintInColor.black('Black', end=' ')
        if colors[i]=='Green':
            PrintInColor.green('Green', end=' ')
        if colors[i]=='White':
            PrintInColor.white('White', end=' ')
```
